```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from glob import glob
import os

from scf.qelattice import get_qel
```

```python
path2target = '/Users/y1u0d2/desktop/Lab/result/qe/Si/mp-slab/coord'
dirs = glob(f'{path2target}/0*')
```

```python
all_scf_dirs = []

for d in dirs:
    scf_dirs = glob(f'{d}/result/scf*')
    all_scf_dirs += scf_dirs
```

```python
coord_list = []
for d in all_scf_dirs:
    try:
        qel = get_qel(d)
        coord_df = pd.DataFrame(data=qel.get_coord(), columns=['fx', 'fy', 'fz'])
        coord_list.append(coord_df)
    except:
        print(d)
        continue
```

```python
all_coord_df = pd.concat(coord_list)
```

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)

ax.scatter(all_coord_df['fx'], all_coord_df['fz'])
ax.set_xlim(0, 10.9375)
ax.set_ylim(0, 32.6159)
```
