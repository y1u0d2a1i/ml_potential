```python
with open(os.path.join(path2target, 'input.data'), mode='r') as f:
    l_strip = [s.strip() for s in f.readlines()]
    
all_data = list(filter(lambda x: 'comment' in x, l_strip))
```

```python
with open(os.path.join(path2target, 'test.data'), mode='r') as f:
    l_strip = [s.strip() for s in f.readlines()]
    
test_data = list(filter(lambda x: 'comment' in x, l_strip))
```

```python
with open(os.path.join(path2target, 'train.data'), mode='r') as f:
    l_strip = [s.strip() for s in f.readlines()]
    
train_data = list(filter(lambda x: 'comment' in x, l_strip))
```

```python
all_df = pd.DataFrame(data=all_data, columns=['comment'])
```

```python
all_df = all_df.reset_index()
all_df = all_df.rename(columns={'index': 'structure_idx'})
```

```python
test_df = pd.DataFrame(data=test_data, columns=['comment'])
train_df = pd.DataFrame(data=train_data, columns=['comment'])
```

```python
test_structure_ids = pd.merge(all_df, test_df, on='comment', how='right')
train_structure_ids = pd.merge(all_df, train_df, on='comment', how='right')
```

```python
def get_materials_id_from_comment(val):
    splitted_path = val.split('/')
    id = list(filter(lambda x: 'mp-' in x, splitted_path))[0]
    return id

test_structure_ids['materials_id'] = test_structure_ids['comment'].map(get_materials_id_from_comment)
train_structure_ids['materials_id'] = train_structure_ids['comment'].map(get_materials_id_from_comment)
```
