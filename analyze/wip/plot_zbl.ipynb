```python
import numpy as np
```

```python
def fai(x):
    return 0.18175 * np.exp(-3.199980 * x) + 0.50986 * np.exp(-0.94229 * x) + 0.28022 * np.exp(-0.40290 * x) + 0.02817 * np.exp(-0.20162 * x)

def zbl(rij, z1, z2):
    e = 1.602e-19
    eps = 8.854e-12
    a = 0.46850 / (z1**0.23 + z2**0.23)
    fai_val = fai(rij/a)
    return (1 / 4 * np.pi * eps) * ((z1 * z1 * e**2) / rij) * fai_val
```

```python
import matplotlib.pyplot as plt

z1 = 13
z2 = 13

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111)

x = np.linspace(1.0, 4, 50)
y = [zbl(rij, z1, z2) for rij in x]

ax.plot(x, y)
```
