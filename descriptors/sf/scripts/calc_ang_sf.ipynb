```python
rcut = 6
ang_param_df = get_sf_ang_params(
    elm=['Si'],
    rcut=rcut,
    nb_param_pairs=3,
    ang_type='angular_wide',
    zetas=[1, 6],
    r_lower=3
)
```

```python
from dscribe.descriptors import ACSF
from ase.io import read

path = all_dirs[0]
structure = read(filename=os.path.join(path, 'scf.in'), format='espresso-in')
species = ['Si', 'Si', 'Si']

g4_params = ang_param_df[['eta', 'zeta', 'lam']].values
columns = [f'eta-{param[0]}_zeta-{param[1]}_lam-{param[2]}' for param in g4_params]

acsf = ACSF(
    species=species,
    rcut=rcut,
    g4_params=g4_params,
    periodic=True,
)
sf_value = acsf.create(system=structure)

columns.insert(0, 'g1')
df = pd.DataFrame(data=sf_value, columns=columns)
```
