```python
import numpy as np
import os
from ase.io import read
from pymatgen.io.ase import AseAtomsAdaptor

def validate_minimum_atomic_diatance(path, rcut):
    n_sites_list = []
    structure = read(os.path.join(path, 'scf.in'), format='espresso-in')
    pg_structure = AseAtomsAdaptor.get_structure(structure)
    all_neighbors = pg_structure.get_all_neighbors(r=rcut)
    for neighbors in all_neighbors:
        if len(neighbors) != 0:
            print(path)
            return False
    return True

rcut = 1.0
all_dirs = list(filter(lambda path: validate_minimum_atomic_diatance(path, rcut=rcut), all_dirs))
print(len(all_dirs))
```
