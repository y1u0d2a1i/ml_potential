```python
result = []
for atoms in all_mlpatoms_sampled:
    try:
        mpid = atoms.structure_id
        n_atom = atoms.n_atoms
        result.append([mpid, n_atom])
    except:
        continue
df = pd.DataFrame(data=result, columns=['mpid', 'n_atom'])
pt = pd.pivot_table(df, index=['mpid', 'n_atom'], aggfunc=len)
```

```python
def get_deepmd_data(all_atoms, path2save):
    box, coord, energy, force = [], [], [], []

    for atoms in all_atoms:
        box.append(np.ravel(atoms.cell))
        coord.append(np.ravel(atoms.coord))
        energy.append(np.ravel(atoms.energy))
        force.append(np.ravel(atoms.force))

    box, coord, energy, force = np.array(box), np.array(coord), np.array(energy), np.array(force)

    np.save(os.path.join(path2save, 'box'), box)
    np.save(os.path.join(path2save, 'coord'), coord)
    np.save(os.path.join(path2save, 'energy'), energy)
    np.save(os.path.join(path2save, 'force'), force)
```

```python
mlpatom_dict = []

for atoms in all_mlpatoms_sampled:
    try:
        mlpatom_dict.append(
            {
                'structure_id': atoms.structure_id,
                'n_atoms': atoms.n_atoms,
                'obj': atoms
            }
        )
    except:
        print(atoms.coord)
        continue
```

```python
import random

train_fraction = 0.9
path2root = '/Users/y1u0d2/desktop/Lab/result/deepmd/Si/35/data/'
path2train = os.path.join(path2root, 'train')
path2test = os.path.join(path2root, 'test')

for (structure_id, n_atoms) in pt.index:

    mlpatoms_for_each_structure_id_and_n_atoms = [d['obj'] for d in mlpatom_dict if d['structure_id'] == structure_id and d['n_atoms'] == n_atoms]
    random.shuffle(mlpatoms_for_each_structure_id_and_n_atoms)

    print(f'structure_id: {structure_id}, n_atoms: {n_atoms}, n_data: {len(mlpatoms_for_each_structure_id_and_n_atoms)}')

    # split data
    train_structure = mlpatoms_for_each_structure_id_and_n_atoms[:int(len(mlpatoms_for_each_structure_id_and_n_atoms) * train_fraction)]
    test_structure = mlpatoms_for_each_structure_id_and_n_atoms[int(len(mlpatoms_for_each_structure_id_and_n_atoms) * train_fraction):]

    # create each mpid dir (train)
    if not os.path.exists(path2train):
        os.mkdir(path2train)

    path2target = os.path.join(path2train, f'{structure_id}_{n_atoms}')
    if not os.path.exists(path2target):
        os.mkdir(path2target)
        path2set = os.path.join(path2target, 'set.000')
        os.mkdir(path2set)
        get_deepmd_data(train_structure, path2set)
        # create type.raw
        with open(os.path.join(path2target, 'type.raw'), mode='w') as f:
            for i in range(n_atoms):
                f.write('0\n')
        # create type_map.raw
        with open(os.path.join(path2target, 'type_map.raw'), mode='w') as f:
            for i in range(1):
                f.write('Si')
    
    # create each mpid dir (test)
    if not os.path.exists(path2test):
        os.mkdir(path2test)

    path2target = os.path.join(path2test, f'{structure_id}_{n_atoms}')
    if not os.path.exists(path2target):
        os.mkdir(path2target)
        path2set = os.path.join(path2target, 'set.000')
        os.mkdir(path2set)
        get_deepmd_data(test_structure, path2set)
        # create type.raw
        with open(os.path.join(path2target, 'type.raw'), mode='w') as f:
            for i in range(n_atoms):
                f.write('0\n')
        # create type_map.raw
        with open(os.path.join(path2target, 'type_map.raw'), mode='w') as f:
            for i in range(1):
                f.write('Si')
        
    

    # get_deepmd_data(test_structure, path2test)
    print(f'num total: {len(mlpatoms_for_each_structure_id_and_n_atoms)}')
    print(f'num train: {len(train_structure)}')
    print(f'num test: {len(test_structure)}')
```
